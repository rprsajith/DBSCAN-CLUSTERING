```python
x=df.iloc[:,3:5].values
x
```

```python
from sklearn.cluster import DBSCAN
db=DBSCAN(eps=3,min_samples=4,metric='euclidean')
text = db.fit(x)
text
```

```python
label = text.labels_
print(label)
```

```python
n_clustering = len(set(label))-(1 if -1 in label else 0)
print('number of clusers:' , n_clustering)
```

```python
n_clustering
```

```python
model=db.fit_predict(x)
plt.figure(figsize=(15,7))
plt.scatter(x[model==0,0],x[model==0,1],color="red",label="cluster 0",s=50)
plt.scatter(x[model==1,0],x[model==1,1],color="orange",label="cluster 1",s=50)
plt.scatter(x[model==2,0],x[model==2,1],color="black",label="cluster 2",s=50)
plt.scatter(x[model==3,0],x[model==3,1],color="blue",label="cluster 3",s=50)
plt.scatter(x[model==4,0],x[model==4,1],color="brown",label="cluster 4",s=50)
plt.scatter(x[model==5,0],x[model==5,1],color="pink",label="cluster 5",s=50)
plt.scatter(x[model==6,0],x[model==6,1],color="grey",label="cluster 6",s=50)
plt.scatter(x[model==7,0],x[model==7,1],color="purple",label="cluster 7",s=50)
plt.scatter(x[model==8,0],x[model==8,1],color="yellow",label="cluster 8",s=50)
plt.show()
```
